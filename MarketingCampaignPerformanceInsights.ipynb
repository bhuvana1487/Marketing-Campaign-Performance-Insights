# Marketing Campaign Performance Insights

**Load the Dataset**

```python
import pandas as pd

# Load CSV file into DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/ArchanaInsights/Datasets/main/marketing_campaign.csv')
#print(df)
#type(df)
```

**Exploratory Data Analysis (EDA) and Visualization**
*Campaign Performance*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

#  scatter plot to visualize the relationship between Acquisition_Cost and ROI
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Acquisition_Cost', y='ROI', hue='Campaign_Type')
plt.title('Acquisition Cost vs ROI')
plt.xlabel('Acquisition Cost')
plt.ylabel('ROI')
plt.legend(title='Campaign Type')
plt.tight_layout()
plt.show()
```

```python
# bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Channel_Used', y='Conversion_Rate', hue='Campaign_Type', errorbar=None, estimator='mean')
plt.xticks(rotation=45)
plt.title('Average Conversion Rate by Channel and Campaign Type')
plt.ylabel('Avg Conversion Rate')
plt.tight_layout()
plt.show()
```

```python
# box plot to visualize the distribution of Engagement_Score across different Campaign_Type
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Campaign_Type', y='Engagement_Score', hue='Campaign_Type', palette='Set3', legend=False)
plt.title('Engagement Score Distribution by Campaign Type')
plt.xlabel('Campaign Type')
plt.ylabel('Engagement Score')
plt.tight_layout()
plt.show()
```

```python
# average ROI by Company using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Company', y='ROI', errorbar=None)
plt.xticks(rotation=45)
plt.title('Average ROI by Company')
plt.ylabel('Avg ROI')
plt.tight_layout()
plt.show()
```

```python
#Engagement_Score and Conversion_Rate using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Engagement_Score', 'Conversion_Rate']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

 **Customer Segmentation**

```python
#count plot to visualize the distribution of Target_Audience.
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Target_Audience', order=df['Target_Audience'].value_counts().index, hue='Target_Audience', palette='pastel', legend=False)
plt.xticks(rotation=45)
plt.title('Distribution of Target Audience')
plt.xlabel('Target Audience')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Customer_Segment has the highest Conversion_Rate for each Language using a bar chart
top_segments = df.groupby(['Language', 'Customer_Segment'])['Conversion_Rate'].mean().reset_index()
top_segments = top_segments.sort_values(['Language', 'Conversion_Rate'], ascending=[True, False])
top_segments = top_segments.groupby('Language').first().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=top_segments, x='Language', y='Conversion_Rate', hue='Customer_Segment', dodge=False, palette='deep')
plt.title('Top Customer Segment by Conversion Rate for Each Language')
plt.ylabel('Conversion Rate')
plt.tight_layout()
plt.show()
```

```python
# distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used, using a box plot

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Customer_Segment', y='Acquisition_Cost', hue='Channel_Used')
plt.title('Acquisition Cost by Customer Segment and Channel Used')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#average Conversion_Rate by Language using a bar chart

plt.figure(figsize=(8, 6))
avg_conversion_lang = df.groupby('Language')['Conversion_Rate'].mean().reset_index().sort_values(by='Conversion_Rate', ascending=False)

sns.barplot(data=avg_conversion_lang, x='Language', y='Conversion_Rate', palette='coolwarm', hue='Language')
plt.title('Average Conversion Rate by Language')
plt.ylabel('Avg Conversion Rate')
plt.tight_layout()
plt.show()
```

**Channel Effectiveness**

```python
# Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Channel_Used', y='Engagement_Score', hue='Campaign_Type', estimator='mean', errorbar=None,
    palette='Set2')
plt.title('Average Engagement Score by Channel and Campaign Type')
plt.ylabel('Avg Engagement Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# distribution of total ROI across different Channels_Used using a pie chart
roi_by_channel = df.groupby('Channel_Used')['ROI'].sum()

plt.figure(figsize=(8, 8))
plt.pie(roi_by_channel, labels=roi_by_channel.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Total ROI Distribution by Channel Used')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
#  scatter plot to show the relationship between Clicks and Impressions for each Channel_Used
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Impressions', y='Clicks', hue='Channel_Used', palette='Set1')
plt.title('Clicks vs Impressions by Channel')
plt.xlabel('Impressions')
plt.ylabel('Clicks')
plt.tight_layout()
plt.show()
```

** Time-Base Analysis**

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
#distribution of Duration using a histogram

plt.figure(figsize=(8, 6))
sns.histplot(df['Duration'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Campaign Duration')
plt.xlabel('Duration (Days)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#  overall Conversion_Rate has changed over Date for each Company using a line chart

conversion_trend = df.groupby(['Date', 'Company'])['Conversion_Rate'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=conversion_trend, x='Date', y='Conversion_Rate', hue='Company')
plt.title('Conversion Rate Trend Over Time by Company')
plt.ylabel('Average Conversion Rate')
plt.tight_layout()
plt.show()
```

```python
# trend of Engagement_Scoreover Date with a line chart
engagement_trend = df.groupby('Date')['Engagement_Score'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=engagement_trend, x='Date', y='Engagement_Score', color='green')
plt.title('Engagement Score Trend Over Time')
plt.ylabel('Average Engagement Score')
plt.tight_layout()
plt.show()
```

Geographic Analysis

```python
#  location has the highest Acquisition_Cost using a bar chart

acq_cost_by_location = df.groupby('Location')['Acquisition_Cost'].mean().sort_values(ascending=False).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=acq_cost_by_location, x='Location', y='Acquisition_Cost', hue='Location', palette='rocket')
plt.title('Average Acquisition Cost by Location')
plt.xticks(rotation=45)
#plt.legend().remove()
plt.tight_layout()
plt.show()
```

```python
# Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Location', y='Conversion_Rate', hue='Target_Audience', errorbar=None)
plt.title('Conversion Rate by Location and Target Audience')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#  proportion of ROI by Location using a pie chart

roi_by_location = df.groupby('Location')['ROI'].sum()

plt.figure(figsize=(8, 8))
plt.pie(roi_by_location, labels=roi_by_location.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Total ROI Distribution by Location')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
